```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\study\Book-recommnder\dylanjcastillo\7k-books-with-metadata\versions\3\books.csv")
df
```

```python
df.info()
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.show()
```

```python
 import numpy as np

 df["miss_desc"] = np.where(df["description"].isna(),1,0)
 df["book_age"] = 2025-df["published_year"]
```

```python
corr_mat =df[["book_age","miss_desc",'num_pages',"average_rating"]].corr(method="pearson")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(corr_mat,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
df_filled =df[~(df['description'].isna())&
~df["num_pages"].isna()&
~df["average_rating"].isna()&
~df["published_year"].isna()]
```

```python
df_filled.info()
```

```python
df_filled["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
df_filled["words_desc_count"]=df_filled["description"].str.split().str.len()
df_filled
```

```python
df_filled.loc[df_filled["words_desc_count"].between(1,5),"description"]
```

```python
df_filled.loc[df_filled["words_desc_count"].between(25,34),"description"]
```

```python
df_filter = df_filled[df_filled["words_desc_count"]>=25]
df_filter
```

```python
df_filter["title_&_desc"] = np.where(df_filter["subtitle"].isna(),df_filter["title"],df_filter[["title",'subtitle']].astype(str).agg("".join,axis=1))
```

```python
df_filter["tagged_description"] =df_filter[["isbn13","description"]].astype(str).agg("".join,axis=1)
```

```python
df_filter
```

```python
df_filter.drop(["subtitle","miss_desc","book_age","words_desc_count"],axis=1).to_csv("df_cleaned.csv",index=False)
```
